```python
everything={} 
numbers = 0

#открывание файла
with open('Проект_группаAL_Гагарина_Решетюк_Торгомян.csv', 'r', encoding = 'utf8') as project: 
    for line in project: #чертовa экселькa
        names = line.strip().split(',') 
        if numbers == 0: 
            for name in names:
                everything[name.strip('"')]=[] 
        else:
            keys = list(everything.keys())
            for i in range(len(keys)):
                if i == 0:
                    everything[keys[i]].append(names[i].strip('"'))
                else:
                    everything[keys[i]].append(round(float(names[i].strip('"')), 2))
        numbers += 1 
    print(everything)
```

```python
minimal_numbers = [] 
minimal_names = [] 

#минимальное значение
def minimal(lol): #без осуждений за лол у меня мозг не генерирует новые названия
    keys = list(lol.keys())
    for i in range(len(keys)):
        if i != 0:
            minimal_numbers.append(min(lol[keys[i]])) #почему нельзя все сделать в экселе??
        for number in minimal_numbers:
            for n in lol[keys[i]]:
                if number == n:
                    minimal_names.append(lol[keys[0]][lol[keys[i]].index(n)])
        if i!=0:
            print(f'Минимальное значение по параметру "{keys[i]}" - {minimal_numbers[i-1]}, {minimal_names[i-1]}') 
            
minimal(everything)
print(minimal_numbers)
print(minimal_names) #проверка данных
```

```python
maximal_numbers = [] 
maximal_names = []

#максимальное значение
def maximal(lo):
    keys = list(lo.keys())
    for i in range(len(keys)):
        if i != 0:
            maximal_numbers.append(max(lo[keys[i]])) 
        for number in maximal_numbers:
            for n in lo[keys[i]]:
                if number == n:
                    maximal_names.append(lo[keys[0]][lo[keys[i]].index(n)])
        if i!=0:
            print(f'Максимальное значение по параметру "{keys[i]}" - {maximal_numbers[i-1]}, {maximal_names[i-1]}')
            
maximal(everything)
print(maximal_numbers)
print(maximal_names) #опять проверка
```

```python
count_values = len(everything['Субъекты']) 
mean_numbers = [] 

#среднее значение
def mean(lolo): #я сейчас специально загуглила. был мультик "приключения пингвиненка лоло". вот поэтому здесь будет lolo
    keys = list(lolo.keys())[1:] #ГОЛОВУ С ПЛЕЕЕЧ 💀 
    for i in range(len(keys)):
        mean_numbers.append(sum(lolo[keys[i]])/count_values)
        print(f'Среднее значение по параметру "{keys[i]}" - {round(mean_numbers[i], 2)}')
        
mean(everything)
print(mean_numbers, count_values)
```

```python
median_numbers = []

#медианы
def median(lololo): #не будем отходить от традиции
    keys = list(lololo.keys())
    for i in range(len(keys)):
        if i != 0:
            numbers_names = {} 
            for j in range(count_values):
                numbers_names[lololo[keys[i]][j]] = lololo[keys[0]][j] 
            median_numbers.append(sorted(lololo[keys[i]])[int(count_values/2)]) 
            print(f'Медиана по параметру "{keys[i]}" - {median_numbers[i-1]}, {numbers_names[median_numbers[i-1]]}')
            
median(everything)
print(median_numbers) #любимая проверка
```

```python
#квартили по каждому параметру, но уже без субъектов (простите меня великодушно, но это слишком муторный процесс)
upperquartile_numbers = [] #верхние квартили
lowerquartile_numbers = [] #нижние квартили

def quartiles(lolol):
    keys = list(lolol.keys())[1:] #опять рубим головы
    for i in range(len(keys)):
        upperquartile_numbers.append(sorted(lolol[keys[i]])[int(3*count_values/4)])
        lowerquartile_numbers.append(sorted(lolol[keys[i]])[int(count_values/4)])
        print(f'Нижний квартиль по параметру "{keys[i]}" - {lowerquartile_numbers[i]}. Верхний квартиль по этому параметру - {upperquartile_numbers[i]}')
quartiles(everything)
print(upperquartile_numbers)
print(lowerquartile_numbers)
```

```python
#рейтинг: топ-10 по каждому из параметров
with open('rating.txt', 'w', encoding = 'utf8') as r:
    def rating(lololol):
        keys = list(lololol.keys())
        for i in range(len(keys)):
            if i != 0:
                numbers_names = {}
                for j in range(count_values):
                    numbers_names[lololol[keys[i]][j]] = lololol[keys[0]][j] 
                ten_numbers_names = sorted(list(numbers_names.keys()), reverse = True)[:10]
                print(f'Топ-10 субъектов по параметру "{keys[i]}":', file = r)
                k = 1
                for key in ten_numbers_names:
                    print(f'{k} место: {numbers_names[key]} ({key})', file = r)
                    k += 1
                print(file = r)
    rating(everything)
```
